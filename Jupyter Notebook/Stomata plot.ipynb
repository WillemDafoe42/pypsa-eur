```python
import pandas as pd
import scipy
import matplotlib.pyplot as plt
```

```python
# create pandas dataframe from csv file with ";" as separator (wenn du das excel file mit notepad ++ öffnest siehst du dass die 
# einträge in wahrheit nicht durch spalten sondern durch ; getrennt sind)
df = pd.read_csv(r"C:\Users\Willem\Downloads\Example_for_Michaela_1.csv", sep = ";", engine = "python")
display(df)

# create intervals here 
intervals = [0,25,50,75,100,125,150,175,200,225,250,275,300,325,350]

# create pandas series for plotting with counts of intervals
df_sum = pd.DataFrame(df["Median"].groupby(pd.cut(df['Median'], intervals)).count())
    
# plot
fig, ax  = plt.subplots(figsize = (10, 5))
ax = df_sum.loc[:,"Median"].plot.bar(stacked = True, 
                                   ax = ax,  
                                   color = 'lightgreen',
                                   legend = True, fontsize = 11)


# Gitternetzlinien
ax.grid(color = 'grey', linestyle = ':', linewidth = 1.0)
# Graphhintergrund
ax.set_facecolor('white')
# y achse
ax.set_ylabel('Frequency',  fontsize = 12)
# x achse
ax.set_xlabel("Stomata count", fontsize=12)
# Diagrammtitel
ax.set_title("Variation in stomata count", fontsize = 14)

# Balkenbeschriftungen
for i, value in enumerate(df_sum.loc[:,"Median"]):
    if value > 7: # wenn der balkenwert > 7 wird der wert darin geschrieben, sonst darüber
        ax.text(i - 0.2 , value - 3, str(value), color='black', fontweight='normal', fontsize = 12)
    elif value > 0:
        ax.text(i - 0.2 , value + 3, str(value), color='black', fontweight='normal', fontsize = 12)
```
